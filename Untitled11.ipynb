```python
from tensorflow.keras.layers import Input, Lambda, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
SIZE = [128, 128]
```

```python
from tensorflow.keras.applications.vgg19 import VGG19
```

```python
vg19 = VGG19(input_shape=SIZE + [3], weights="imagenet", include_top=False)
```

```python
for layer in vg19.layers:
    layer.trainable = False
```

```python
x = Flatten()(vg19.output)
```

```python
prediction = Dense(len(folders), activation="softmax")(x)

modelvg = Model(inputs=vg19.input, outputs=prediction)
```

```python
modelvg.summary()
```

```python
modelvg.compile(loss="categorical_crossentropy", metrics=["accuracy"], optimizer="adam")
```

```python
train_datagen_vg19 = ImageDataGenerator(rescale=1./255)

test_datagen_vg19 = ImageDataGenerator(rescale=1./255)
```

```python

# foreground extraction

import numpy as np
import cv2
from matplotlib import pyplot as plt

# paths = "/content/New Plant Diseases Dataset(Augmented)/train/*"

# for i in paths:
#   image = cv2.imread("")

image= cv2.imread("/content/New Plant Diseases Dataset(Augmented)/train/Tomato___Bacterial_spot/00639d29-2d1a-4fcf-9bd3-a2b3109c74c4___UF.GRC_BS_Lab Leaf 1054.JPG")

plt.imshow(image)
plt.colorbar()
plt.show()
# # creating a simple mask image similar to the loaded image, with the shape and return type
mask = np.zeros(image.shape[:2], np.uint8)

backgroundModel = np.zeros((1, 65), np.float64)
foregroundModel = np.zeros((1, 65), np.float64)

rectangle = (1, 1, 249, 249)

cv2.grabCut(image, mask, rectangle, backgroundModel, foregroundModel,3, cv2.GC_INIT_WITH_RECT)

mask2 = np.where((mask == 2)|(mask == 0), 0, 1).astype('uint8')

image = image * mask2[:, :, np.newaxis]

# # output segmented image with colorbar
plt.imshow(image)
plt.colorbar()
plt.show()
```

```python
from PIL import Image
import os
import cv2
import tensorflow as tf
from PIL import Image as im
from numpy import asarray
import numpy as np

import numpy as np
from matplotlib import pyplot as plt

for current_directory, sub_directory_names, file_names in os.walk("/content/New Plant Diseases Dataset(Augmented)/train"):
    for name in file_names:
        if name.endswith(".jpg") or name.endswith(".JPG"):
          
          image_path = os.path.join(current_directory, name)
          print(image_path)
          image = cv2.imread(image_path)
          height, width, _ = image.shape
          left_margin_proportion = 0.3
          right_margin_proportion = 0.3
          up_margin_proportion = 0.1
          down_margin_proportion = 0.1

          boundary_rectangle = (
              int(width * left_margin_proportion),
              int(height * up_margin_proportion),
              int(width * (1 - right_margin_proportion)),
              int(height * (1 - down_margin_proportion)),
          )
          gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
          binarized_image =cv2.adaptiveThreshold(gray_image,maxValue=1,adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C,thresholdType=cv2.THRESH_BINARY,blockSize=9,C=7,)
          mask = np.zeros((height, width), np.uint8)
          mask[:] = cv2.GC_PR_BGD
          mask[binarized_image == 0] = cv2.GC_FGD

          background_model = np.zeros((1, 65), np.float64)
          foreground_model = np.zeros((1, 65), np.float64)
          
          if(background_model.any() == 0 or foreground_model.any() == 0):
            continue

          cv2.grabCut(image,mask,boundary_rectangle,background_model,foreground_model,3,cv2.GC_INIT_WITH_MASK,)

          mask2 = np.where((mask == cv2.GC_PR_BGD) | (mask == cv2.GC_BGD), 0, 1).astype("uint8")

          image =image * mask2[:, :, np.newaxis]
          data = im.fromarray(image)
          data.save(image_path)
```

```python
trainning_set_vg19 = train_datagen_vg19.flow_from_directory(train_path,
                                                 target_size=(128, 128),
                                                 batch_size=16,
                                                 class_mode="categorical", shuffle=True)
```

```python
testing_set_vg19 = test_datagen_vg19.flow_from_directory(test_path,
                                                 target_size=(128, 128),
                                                 batch_size=16,
                                                 class_mode="categorical", shuffle=False)
```

```python
import tensorflow as tf


callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)

r_vg19 = modelvg.fit_generator(trainning_set_vg19,
                       validation_data=testing_set_vg19,
                       epochs=15,
                       callbacks=[callback]
                       )
```

```python
import matplotlib.pyplot as plt
accuracy = r_vg19.history['accuracy']
val_accuracy = r_vg19.history['val_accuracy']
loss = r_vg19.history['loss']
val_loss = r_vg19.history['val_loss']
epochs = range(len(accuracy))
plt.title("VGG19")
plt.plot(epochs, accuracy, "b", label="trainning accuracy")
plt.plot(epochs, val_accuracy, "r", label="validation accuracy")
plt.legend()
plt.show()

plt.plot(epochs, loss, "b", label="trainning loss")
plt.plot(epochs, val_loss, "r", label="validation loss")
plt.legend()
plt.show()
```
